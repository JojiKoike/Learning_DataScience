```python
# Practice 4-1
coin_data = np.array([0, 1])
coin_toss_steps = 1000
coin_toss = np.random.choice(coin_data, coin_toss_steps)
for i in range(2):
    p = len(coin_toss[coin_toss == i]) / coin_toss_steps
    print("{0} : {1}".format(i, p))
```
