```python
import requests, zipfile
from io import StringIO
import io

url =  'http://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'
r = requests.get(url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

```python
# HIstogram
plt.hist(student_data_math['absences'])
plt.xlabel('absences')
plt.ylabel('count')
plt.grid(True)
```

```python
# Average, Median, Mode
print("Average : {0}".format(student_data_math['absences'].mean()))
print("Median : {0}".format(student_data_math['absences'].median()))
print("Mode : {0}".format(student_data_math['absences'].mode()))
```

```python
# Deviation, Std. Deviation
print("Variant : {0}".format(student_data_math['absences'].var()))
print("Std. : {0}".format(student_data_math['absences'].std()))
```
