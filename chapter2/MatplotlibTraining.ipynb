```python
import numpy as np
import numpy.random as random
import matplotlib as mpl
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline

# Fix random seed value
random.seed(0)

# x axis
x = np.random.randn(30)
# y axis
y = np.sin(x) + np.random.randn(30)

# Set Graph Area Size
plt.figure(figsize=(20,5))

# Plot Scatter
#plt.plot(x, y, 'o')
plt.scatter(x, y)

# Title
plt.title('Relation Map')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
# Continuous Curve

# Data Range
numpy_data_x = np.arange(1000)

# Generate Random Number and Accumulation
numpy_random_data_y = np.random.randn(1000).cumsum()

plt.figure(figsize=(20,6))
plt.plot(numpy_data_x, numpy_random_data_y, label='Nikei225')
plt.legend()
plt.xlabel('Date Time')
plt.ylabel('Price')
plt.grid(True)
```

```python
plt.figure(figsize=(20,6))

# Graph No.1 
plt.subplot(2,1,1)
x = np.linspace(-10,10,100)
plt.plot(x, np.sin(x))

# Graph No.2
plt.subplot(2,1,2)
y = np.linspace(-10, 10, 100)
plt.plot(y, np.sin(2*y))
```

```python
def my_func(x: float)  -> float:
    return x**2 + 2 * x + 1

x = np.arange(-10, 10)
plt.figure(figsize=(20,6))
plt.plot(x, my_func(x))
plt.grid(True)
```

```python
# Histogram
random.seed(0)
plt.figure(figsize=(20,6))
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins=60, range = (20,80))
plt.grid(True)
```
